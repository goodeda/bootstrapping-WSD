## #KWIC function (key word in context)#: extracting sentences with keywords and select left and right context (10 words as a range))

```python
import re
def KWIC(keyword,texts):
    contexts=[]
    extracts=[]
    for i in texts:
        if keyword in i.split() and re.search(r'^[ ]?\b%s\b|\b%s\b[. ]?$'%(keyword,keyword),i)==None:
            # if there is no context around the key words, then drop it. For example, if the keyword is the first or the last word. 
            contexts.append(i)
    for i,content in enumerate(contexts):
        try:
            context=re.findall('([\w\W ]{2,}) (%s) ([\w\W ]{2,})'%keyword,content)
            #print(i,content)
            extracts.append([context[0][0].split()[-10:],context[0][1].strip(),context[0][2].split()[:10]])# 10-word range
        except:
            continue
    return extracts
```

## #Find pattern function#: discovering sentences according to decision list

```python
def find_pattern(extracts,decision_list):
    extraction_1=[]
    extraction_2=[]
    for sentence in extracts:
        for word in decision_list:# go through the decision list
            if '_'==word[0][-1]: # in form of 'xxx_', before keyword
                if word[0].replace('_','')==sentence[0][-1]:
                    if word[1][0]>word[1][1]: #if x>y, the word sense belongs to type 1
    # since the data structure will be like ('word',[x,y,z]) 
    #where x is the sense 1 count, y is the sense 2 count, z is the score
                        extraction_1.append(sentence)# otherwise, it's type 2
                        break #once it is found, the loop can be ended and start the next sentence
                    else:
                        extraction_2.append(sentence)
                        break
            elif '_'==word[0][0]:#in form of '_xxx', after keyword
                if word[0].replace('_','')==sentence[2][0]:
                    if word[1][0]>word[1][1]:
                        extraction_1.append(sentence)
                        break
                    else:
                        extraction_2.append(sentence)
                        break
            else:
                if word[0] in sentence[0][:-1] or word[0] in sentence[2][1:]:# 'xxx' within (+-)10-word window
                    if word[1][0]>word[1][1]:# same as above
                        extraction_1.append(sentence)
                        break
                    else:
                        extraction_2.append(sentence)
                        break
        continue# if it can't be decided with decision list, then leave it aside
    print(len(extraction_1),len(extraction_2))
    return extraction_1,extraction_2,len(extraction_1),len(extraction_2)
```

## #Count score function#: store the collocation in '_sense_', and bag of words in 'sense'. The data structure is: 'pattern_word': [x,y,z]. x is the number of appearing in sense 1 sentence, y is the number in sense 2 sentence and z is calculated by $abs(log(\frac{Pr(sense_{1}|collocation_{i})}{Pr(sense_{2}|collocation_{i})}))$ and since the denominator of $Pr( )$ is the same so just need to compare the frequency. In order to avoid division by zero, I applied smoothing by adding $\alpha$ on both numerator and denominator.  

```python
import math
def count_score(sense_1,sense_2,alpha):
    feature={'sense':{},'_sense_':{}}
    for window in sense_1:# in the sense 1, it should be added on x
        context_1=window[0][:-1]+window[2][1:]# (+-)10-word range
        for word in context_1:
            if word in feature['sense']:
                feature['sense'][word][0]+=1
            else:
                feature['sense'][word]=[1,0,0]
        if window[0][-1]+'_' in feature['_sense_']:
            feature['_sense_'][window[0][-1]+'_'][0]+=1
        else:
            feature['_sense_'][window[0][-1]+'_']=[1,0,0]
        if '_'+window[2][0] in feature['_sense_']:
            feature['_sense_']['_'+window[2][0]][0]+=1
        else:
            feature['_sense_']['_'+window[2][0]]=[1,0,0]
    for window in sense_2:# in the sense 2, it should be added on y
        context_2=window[0][:-1]+window[2][1:]
        for word in context_2:
            if word in feature['sense']:
                feature['sense'][word][1]+=1
            else:
                feature['sense'][word]=[0,1,0]
        if window[0][-1]+'_' in feature['_sense_']:
            feature['_sense_'][window[0][-1]+'_'][1]+=1
        else:
            feature['_sense_'][window[0][-1]+'_']=[0,1,0]#1st digit--> sense 1, 2nd digit-->sense 2, 3rd digit-->rank_score
        if '_'+window[2][0] in feature['_sense_']:
            feature['_sense_']['_'+window[2][0]][1]+=1
        else:
            feature['_sense_']['_'+window[2][0]]=[0,1,0]
    for collocation in feature:# calculate the value of z
        if '_' in collocation:
            for word in feature[collocation]:
                feature[collocation][word][2]=abs(math.log((feature['_sense_'].get(word,[0,0,0])[0]+alpha)/(feature['_sense_'].get(word,[0,0,0])[1]+alpha)))
        else:    
            for word in feature[collocation]:
                feature[collocation][word][2]=abs(math.log((feature['sense'].get(word,[0,0,0])[0]+alpha)/(feature['sense'].get(word,[0,0,0])[1]+alpha)))
    return feature
```

## #Add new keywords function#: select newly discovered words, and add them into the decision list. We should set up a threshold to filter out some words with lower confidence level. 

```python
def add_new_keywords(feature,threshold):
    total_list=[]
    for keywords in feature.values():
        total_list.extend(list(keywords.items()))
    #new_add_sense=[keys[0] for keys in total_list if keys[1][2]>2.5 and abs(keys[1][0]-keys[1][1])>=2]#hyperparameter
    #print(total_list)
    selected_ranking=sorted([keys for keys in total_list if keys[1][2]>threshold and abs(keys[1][0]-keys[1][1])>=2],key=lambda x:x[1][2],reverse=True)
    #print(selected_ranking)
    return selected_ranking
#print(len(pattern_words['factory']),len(pattern_words['factory']))
```

# randomly select instances and store the result data in excel format and manually check it

```python
import numpy as np
import random
def random_select(text,keyword,N):
    random.seed(11)
    random_index=random.sample(range(0,len(text)),N)#generating ramdom index
    random_select=[text[i] for i in random_index]
    return random_select
def assign_sense(context_extraction,sense1_len,sense2_len,decision_list):
    sense_list=[]
    for sentence in context_extraction:
        assigned=False
        for word in decision_list:
            #print(word)
            if '_' in word[0] and '_'==word[0][-1]: # if word is the left word next to keyword
                if word[0].replace('_','')==sentence[0][-1]:
                    sense_list.append(0 if word[1][0]>word[1][1] else 1)
                    assigned=True
                    break
            elif '_' in word[0] and '_'==word[0][0]:# if the word is the right word next to keyword
                if word[0].replace('_','')==sentence[2][0]:
                    sense_list.append(0 if word[1][0]>word[1][1] else 1)
                    assigned=True
                    break
            else:
                if word[0] in sentence[0][:-1] or word[0] in sentence[2][1:]:# if the word in 2~10-word scope
                    sense_list.append(0 if word[1][0]>word[1][1] else 1)
                    assigned=True
                    break
        if not assigned:# if the decision list can not decide, assign a sense according to the probabilities.
            # the probabilities are calculated by the number of decided instance of sense 1 and 2.
            sense_list.append(int(np.random.choice([0,1],size=1,p=[sense1_len/(sense1_len+sense2_len),sense2_len/(sense1_len+sense2_len)])))
    assert len(sense_list)==len(context_extraction)
    return sense_list
```

```python
import random
def random_select_auto(text,keyword,N):
    random.seed(11)
    random_index=random.sample(range(0,len(text)),N)#generating ramdom index
    random_select=[text[i] for i in random_index]
    return random_select,random_index
def accuracy_cal(predict,true):
    score=0
    assert len(predict)==len(true)
    for i in range(0,len(predict)):
        if predict[i]==true[i]:
            score+=1
    return 'accuracy:%.3f %%'%(100*score/len(predict))
```
